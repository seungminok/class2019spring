# 모든 링크의 텍스트와 주소 가져오기

bsObject 전체에서 a로 시작되는 태그 <a href =>

```python
for link in bsObject.find_all('a'):
    print(link.get_text(), link.get('href'))
```

```python
# Create a file to write to, add headers row
f = csv.writer(open('crawling.csv', 'w'))
f.writerow(['Name', 'Link'])

for link in bsObject.find_all('a'):
    names = link.text.strip()
    links = link.get('href')

    f.writerow([names, links])
```

>dt태그에서 select 라는 거? a라는 태그를 찾아라. 찾아서, 그것의 첫번째 꺼 [0] 거기서 get text 다 버리고 순수한 말만 가져와. 그러면 책 제목을 가져올 수 있다.

>print(dt.next_sibling.next_sibling)

>print(dt.next_sibling.next_sibling.select('a')) a태그가 들어있는 거 다 받아줌. 쉼표로 이어진 리스트로 받아짐.  
>
>publisher은 dd전체에서 get text()를 써버리면, 모든 꺽쇠를 다 없애버림. 그 상태에서, split을 써버리면[-1] 젤마지막에 있는 것을 쓰겠다. 아마도 출판사이름 책구루 일 것이다. 
>
>그걸 print해서, 1, title, author, publisher
>
>dt.select('a') [0].get_text a만 포함한 걸 printout
>
>split default가 띄어쓰기로 쪼개는 거 print(dt.next_sibling.next_sibling.get_text().split()[3])하면 책구루가 나와야 됨. 3 대신 -1 적는게 더 편리 

```python
for n in range(0,25):
    dt = bsObject.find("dt",{"id":"book_title_"+str(n)})
    title = dt.select('a')[0].get_text()
    author = dt.next_sibling.next_sibling.select('a')[0].get_text()
    publisher = dt.next_sibling.next_sibling.get_text().split()[-1]
    print(n+1, title, author, publisher)
```

>open('crawling.csv','w') as f:
>
>엑셀파일에 정리할 수 있다.
>
>writerow 젤 첫줄에 뭐가 나오게 할 건지.

```python
# Create a file to write to, add headers row
with open('crawling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'author', 'publisher'])
        
    for n in range(0,25):
        dt = bsObject.find("dt", {"id":"book_title_"+str(n)})
        title = dt.select('a')[0].get_text()
        author = dt.next_sibling.next_sibling.select('a')[0].get_text()
        publisher = dt.next_sibling.next_sibling.get_text().split()[-1]
        writer.writerow([n+1, title, author, publisher])
```
